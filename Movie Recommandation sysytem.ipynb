```python
credit_colunm_rename=credits.rename(index=str,columns={'movie_id':'id'})
merge_movies=movies.merge(credit_colunm_rename,on ='id')
merge_movies.head()
```

```python
movies=merge_movies.drop(['homepage','crew','title_x', 'title_y', 'status','production_countries'],axis='columns')
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
        return L
        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies['production_companies']=movies['production_companies'].apply(convert)
movies['spoken_languages']=movies['spoken_languages'].apply(convert)
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tagline']=movies['tagline'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']     
```

```python
movies.head()
```

```python
new_df=movies[['id','original_title','tags']]
```

```python
new_df
```

```python
# Step 1: Inspect the Column
print(new_df['tags'].apply(type).unique())
non_iterable_elements = new_df['tags'][new_df['tags'].apply(lambda x: not isinstance(x, (list, tuple)))]
print(non_iterable_elements)

# Step 2: Convert to Iterables (if necessary)
new_df['tags'] = new_df['tags'].apply(lambda x: x if isinstance(x, (list, tuple)) else [x])

# Step 3: Apply the Join Function
new_df['tags'] = new_df['tags'].apply(lambda x: ",".join(map(str, x)))

# Display the DataFrame to verify the result
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())

```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)     
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# Vactorisation of tags data so that convert text into a array matrics

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

# using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])[1:11]
```

```python
def recommend(movies):
    movie_index=new_df[new_df['original_title']==movies].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].original_title)
    return
```

```python
recommend('Batman Begins')
```
